### Outlier Detection with Boxplots

```python
# Identify numeric columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Check shape before removing outliers
print(f"Shape before outlier removal: {df.shape}")
```

```python
print(df.columns)
```

```python
#boxplot for checking outliers in data
plt.figure(figsize=(20, 15))

for i, col in enumerate(num_cols, 1):
    plt.subplot((len(num_cols) // 4) + 1, 4, i)  # auto adjust rows
    sns.boxplot(x=df[col], color='skyblue')
    plt.title(f"Boxplot of {col}")

plt.tight_layout()
plt.show()
```

```python
# Removing outliers using IQR method
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df[num_cols] < (Q1 - 1.5 * IQR)) | (df[num_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]

print(f"Data shape after outlier removal: {df.shape}")
```

### Clustering Model Building

### KMeans

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
# Select Features
features = [
    'Age', 'Income', 'Total_Spending',
    'NumWebPurchases', 'NumDealsPurchases', 'NumStorePurchases'
]
X = df[features].copy()
```

```python
# Standardize Data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Function to find optimal k using Elbow & Silhouette
def find_optimal_k(data, max_k=10):
    wcss, silhouette_scores = [], []
    for k in range(2, max_k+1):
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        labels = kmeans.fit_predict(data)
        wcss.append(kmeans.inertia_)
        silhouette_scores.append(silhouette_score(data, labels))
    return wcss, silhouette_scores

# Calculate metrics
wcss, sil_scores = find_optimal_k(X_scaled)
```

```python
X_scaled
```

```python
# Plot Elbow & Silhouette curves
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].plot(range(2, 11), wcss, marker='o')
ax[0].set_title('Elbow Method (WCSS)')
ax[0].set_xlabel('Number of Clusters (k)')
ax[0].set_ylabel('WCSS')

ax[1].plot(range(2, 11), sil_scores, marker='o', color='green')
ax[1].set_title('Silhouette Scores')
ax[1].set_xlabel('Number of Clusters (k)')
ax[1].set_ylabel('Silhouette Score')
plt.show()
```

```python
# Fit K-Means with chosen k
optimal_k = 4  # Based on silhouette & elbow results
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df['KMeans_Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
# Evaluate model
print("K-Means Silhouette Score:", silhouette_score(X_scaled, df['KMeans_Cluster']))
print("K-Means Davies-Bouldin Score:", davies_bouldin_score(X_scaled, df['KMeans_Cluster']))
```

```python
# Cluster value counts
print("\nCluster Counts (K-Means):")
print(df['KMeans_Cluster'].value_counts())
```

```python
# Cluster Profiles
cluster_profile = df.groupby('KMeans_Cluster')[features].mean()
cluster_profile
```

```python
# PCA Visualization (Bright Colors)
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=X_pca[:, 0], y=X_pca[:, 1],
    hue=df['KMeans_Cluster'],
    palette=sns.color_palette("bright", optimal_k),
    s=80, edgecolor='black'
)
plt.title(f'K-Means Clustering (k={optimal_k}) - PCA Projection', fontsize=14)
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(title='Cluster')
plt.show()
```

### Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Using Ward's Method(Linkage)
plt.figure(figsize=(10, 5))
linkage_matrix = linkage(X, method='ward')
dendrogram(linkage_matrix, truncate_mode='level', p=5)
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

```python
# Trying different cluster numbers
results = {}
sil_scores_hier = []

# Looping Cluster numbers (2-10 clusters)
for k in range(2, 11):
    hier = AgglomerativeClustering(n_clusters=k).fit(X)
    sil_scores_hier.append(silhouette_score(X, hier.labels_))

# Plotting Silhouette  scores
plt.figure(figsize=(6,4))
plt.plot(range(2, 11), sil_scores_hier, 'go-')
plt.xlabel("Number of clusters")
plt.ylabel("Silhouette Score")
plt.title("Hierarchical Clustering Silhouette Scores")
plt.show()
```

```python
# Finding Best K
best_k_hier = np.argmax(sil_scores_hier) + 2
print(f"Best k for Hierarchical: {best_k_hier}")

hier = AgglomerativeClustering(n_clusters=best_k_hier).fit(X)
labels_hier = hier.labels_
results['Hierarchical'] = silhouette_score(X, labels_hier)

plt.figure(figsize=(6,4))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=labels_hier, palette="tab10")
plt.title(f"Hierarchical Clustering (k={best_k_hier})")
plt.show()
```

### DBSCAN

### We use the k-distance graph and KneeLocator to estimate a good eps.

```python
from sklearn.neighbors import NearestNeighbors
from kneed import KneeLocator
import numpy as np
import matplotlib.pyplot as plt

min_samples = 5  # initial guess

# Compute k-distances
neighbors = NearestNeighbors(n_neighbors=min_samples)
neighbors_fit = neighbors.fit(X_scaled)
distances, indices = neighbors_fit.kneighbors(X_scaled)

# Sort k-distances
distances = np.sort(distances[:, min_samples-1])

# Find "elbow" automatically
kneedle = KneeLocator(range(len(distances)), distances, S=1.0, curve="convex", direction="increasing")
eps_auto = distances[kneedle.knee]

# Plot
plt.figure(figsize=(6,4))
plt.plot(distances)
plt.axhline(y=eps_auto, color='r', linestyle='--', label=f"eps={eps_auto:.2f}")
plt.title("Automatic eps detection")
plt.xlabel("Points sorted by distance")
plt.ylabel(f"{min_samples}-NN distance")
plt.legend()
plt.show()

print("Auto-selected eps:", eps_auto)
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=eps_auto, min_samples=min_samples).fit(X_scaled)
labels = dbscan.labels_
```

```python
from sklearn.metrics import silhouette_score

if len(set(labels)) > 1 and not (set(labels) == {-1}):
    sil = silhouette_score(X_scaled, labels)
    print("Silhouette Score:", sil)
else:
    print("DBSCAN found only 1 cluster (or all noise).")
```

```python
import seaborn as sns
from sklearn.decomposition import PCA

X_pca = PCA(n_components=2).fit_transform(X_scaled)

plt.figure(figsize=(6,4))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=labels, palette="tab10")
plt.title(f"DBSCAN Clustering (eps={eps_auto:.2f}, min_samples={min_samples})")
plt.show()
```

## Mean Shift

```python
from sklearn.cluster import MeanShift, estimate_bandwidth

bandwidth = estimate_bandwidth(X_scaled, quantile=0.2, n_samples=500)
meanshift = MeanShift(bandwidth=bandwidth).fit(X_scaled)
labels_meanshift = meanshift.labels_

if len(set(labels_meanshift)) > 1:
    results['MeanShift'] = {
        "Best Params": f"bandwidth={bandwidth:.2f}",
        "Silhouette": silhouette_score(X_scaled, labels_meanshift),
        "DBI": davies_bouldin_score(X_scaled, labels_meanshift)
    }
else:
    results['MeanShift'] = {
        "Best Params": f"bandwidth={bandwidth:.2f}",
        "Silhouette": "N/A (only 1 cluster)",
        "DBI": "N/A"
    }

results
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=labels_meanshift, palette="tab10")
plt.title("Mean Shift Clustering")
plt.show()
```

```python
print("\nSilhouette Scores:")
for model, score in results.items():
    print(f"{model}: {score}")
```
